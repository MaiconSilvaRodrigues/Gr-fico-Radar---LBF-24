```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Função para fazer o web scraping e retornar um DataFrame com os dados
def scrape_data(url, columns):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        linhas = soup.select(columns['linha'])

        dados = []

        for linha in linhas:
            item = {}
            for coluna, classe in columns['colunas'].items():
                if coluna == 'Nome do Atleta':
                    nome = linha.find('td', class_=classe).text.strip()
                    item[coluna] = nome
                else:
                    valor = linha.find('td', class_=classe).text.strip()
                    item[coluna] = valor
            dados.append(item)

        df = pd.DataFrame(dados)
        return df
    else:
        print('Falha ao obter a página:', response.status_code)
        return None

# URLs e configurações para cada tipo de estatística
urls = {
    'pontos': 'https://lbf.com.br/estatisticas/pontos/?a=1&b=23&c=0&d=0&e=1&f=0&g=0&h=1#',
    'arremessos': 'https://lbf.com.br/estatisticas/arremessos/?a=1&b=23&c=0&d=0&e=1&f=0&g=0&h=1',
    'rebotes': 'https://lbf.com.br/estatisticas/rebotes/?a=1&b=23&c=0&d=0&e=1&f=0&g=0&h=1',
    'assistencias': 'https://lbf.com.br/estatisticas/assistencias/?a=1&b=23&c=0&d=0&e=1&f=0&g=0&h=1',
    'roubos': 'https://lbf.com.br/estatisticas/roubos/?a=1&b=23&c=0&d=0&e=1&f=0&g=0&h=1',
    'tocos': 'https://lbf.com.br/estatisticas/tocos/?a=1&b=23&c=0&d=0&e=1&f=0&g=0&h=1'
}

# Configurações para cada tipo de estatística
configuracoes = {
    'pontos': {
        'linha': "#tablePontosIndividuais > tbody > tr",
        'colunas': {
            'Posição': 'cell_posicao',
            'Nome do Atleta': 'cell_atleta',
            'Jogos': 'cell_jogos',
            'Minutos': 'cell_minutos',
            'Pontos': 'cell_pontos',
            'Arremessos 3P': 'cell_arremessos_3p',
            'Arremessos 2P': 'cell_arremessos_2p',
            'Arremessos LL': 'cell_arremessos_ll'
        }
    },
    'arremessos': {
        'linha': "#tableIndividual > div.tableEstatisticasInternoNova > table > tbody > tr",
        'colunas': {
            'Nome do Atleta': 'cell_equipe',
            '2FGM': 'cell_arremessos_3pc',
            '2FGA': 'cell_arremessos_3pt',
            '3FGM': 'cell_arremessos_2pc',
            '3FGA': 'cell_arremessos_2pt',
            'LLT': 'cell_arremessos_llt',
            'LLC': 'cell_arremessos_llc'
        }
    },
    'rebotes': {
        'linha': "#tablePontosIndividuais > tbody > tr",
        'colunas': {
            'Nome do Atleta': 'cell_atleta',
            'Rebotes': 'cell_rebotes',
            'Rebotes Defensivos': 'cell_rebotes_defensivos',
            'Rebotes Ofensivos': 'cell_rebotes_ofensivos'
        }
    },
    'assistencias': {
        'linha': "#tablePontosIndividuais > tbody > tr",
        'colunas': {
            'Nome do Atleta': 'cell_atleta',
            'Assists': 'cell_assistencias',
            'Erros': 'cell_erros',
            'AST/TOV': 'cell_indice'
        }
    },
    'roubos': {
        'linha': "#tablePontosIndividuais > tbody > tr",
        'colunas': {
            'Nome do Atleta': 'cell_atleta',
            'Roubos': 'cell_roubos'
        }
    },
    'tocos': {
        'linha': "#tablePontosIndividuais > tbody > tr",
        'colunas': {
            'Nome do Atleta': 'cell_equipe',
            'Tocos': 'cell_pontos'
        }
    }
}

# Dicionário para armazenar os DataFrames
dataframes = {}

# Scraping e salvando os dados em um dicionário de DataFrames
for tipo, url in urls.items():
    df = scrape_data(url, configuracoes[tipo])
    if df is not None:
        dataframes[tipo] = df

print(dataframes)
```

```python
import re

# Criando um DataFrame consolidado
df_consolidado = dataframes['pontos'].copy()  # Inicializando com os dados de pontos

# Iterando sobre as estatísticas, exceto 'pontos', e combinando os DataFrames
for tipo, df in dataframes.items():
    if tipo != 'pontos':
        df_consolidado = pd.merge(df_consolidado, df, on='Nome do Atleta', how='outer', suffixes=('', f'_{tipo}'))

# Ordenando as colunas do DataFrame consolidado
colunas = ['Nome do Atleta', 'Jogos', 'Minutos', 'Pontos', '2FGM', '2FGA', '3FGM', '3FGA', 'LLT', 'LLC',
          'Rebotes', 'Rebotes Defensivos', 'Rebotes Ofensivos', 'Assists', 'Erros', 'AST/TOV', 'Roubos', 'Tocos']
for tipo in dataframes.keys():
    for coluna in df_consolidado.columns:
        if coluna.endswith(f'_{tipo}'):
            colunas.append(coluna)
df_consolidado = df_consolidado[colunas]
# Renomeando as colunas 'LLC' para 'LLT' e 'LLT' para 'LLC'
df_consolidado = df_consolidado.rename(columns={'LLC': 'LLT', 'LLT': 'LLC', 'Nome do Atleta': 'Atleta',
                                                'Assists': 'Assistências'})
# Criando a coluna 'FGM' como a soma de '2FGM' e '3FGM'
# Convertendo as colunas '2FGM' e '3FGM' para números
df_consolidado['2FGM'] = pd.to_numeric(df_consolidado['2FGM'])
df_consolidado['3FGM'] = pd.to_numeric(df_consolidado['3FGM'])
df_consolidado['2FGA'] = pd.to_numeric(df_consolidado['2FGA'])
df_consolidado['3FGA'] = pd.to_numeric(df_consolidado['3FGA'])



# Criando a coluna 'FGM' como a soma de '2FGM' e '3FGM'
df_consolidado['FGM'] = df_consolidado['2FGM'] + df_consolidado['3FGM']
df_consolidado['FGA'] = df_consolidado['2FGA'] + df_consolidado['3FGA']
df_consolidado['LLC'] = pd.to_numeric(df_consolidado['LLC'], errors='coerce')
df_consolidado['LLT'] = pd.to_numeric(df_consolidado['LLT'], errors='coerce')
# Convertendo as colunas 'Roubos' e 'Tocos' para o tipo de dados numérico
df_consolidado['Roubos'] = pd.to_numeric(df_consolidado['Roubos'], errors='coerce')
df_consolidado['Tocos'] = pd.to_numeric(df_consolidado['Tocos'], errors='coerce')

# Realizando a soma correta das colunas 'Roubos' e 'Tocos'
df_consolidado['STL + BLK'] = df_consolidado['Roubos'] + df_consolidado['Tocos']


# Calculando a coluna 'FG%'
df_consolidado['FG%'] = ((df_consolidado['FGM'] / df_consolidado['FGA']) * 100).round(1)
df_consolidado['2FG%'] = ((df_consolidado['2FGM'] / df_consolidado['2FGA']) * 100).round(1)
df_consolidado['3FG%'] = ((df_consolidado['3FGM'] / df_consolidado['3FGA']) * 100).round(1)
df_consolidado['LL%'] = ((df_consolidado['LLC'] / df_consolidado['LLT']) * 100).round(1)
df_consolidado['FTr'] = ((df_consolidado['LLT'] / df_consolidado['FGA']) * 100).round(1)



# Ordenando o DataFrame pelo nome do jogador de A a Z
df_consolidado = df_consolidado.sort_values(by='Atleta')
# Renomeando as colunas 'Rebotes Defensivos' e 'Rebotes Ofensivos'
df_consolidado = df_consolidado.rename(columns={'Rebotes Defensivos': 'Def Reb', 'Rebotes Ofensivos': 'Off Reb'})

# Exibindo o DataFrame com porcentagens nas colunas '2FG%', '3FG%' e 'FG%'
df_consolidado['2FG%'] = df_consolidado['2FG%'].astype(str) + '%'
df_consolidado['3FG%'] = df_consolidado['3FG%'].astype(str) + '%'
df_consolidado['FG%'] = df_consolidado['FG%'].astype(str) + '%'
df_consolidado['LL%'] = df_consolidado['LL%'].astype(str) + '%'


# Convertendo as colunas 'Minutos' e 'Jogos' para o tipo numérico
# Convertendo as colunas 'Minutos' e 'Jogos' para o tipo numérico
df_consolidado['Minutos'] = pd.to_numeric(df_consolidado['Minutos'])
df_consolidado['Jogos'] = pd.to_numeric(df_consolidado['Jogos'])

# Criando a coluna 'Minutoss' como a multiplicação de 'Minutos' por 'Jogos'
df_consolidado['Minutoss'] = df_consolidado['Minutos'] * df_consolidado['Jogos']

# Removendo a coluna 'Minutos'
df_consolidado = df_consolidado.drop(columns=['Minutos'])
df_consolidado['Atleta'] = df_consolidado['Atleta'].str.strip()

# Removendo a parte entre parênteses da coluna "Atleta" e adicionando um espaço após o símbolo '#'
df_consolidado['Jogadora'] = df_consolidado['Atleta'].apply(lambda x: re.sub(r'\s*\(\w+\)\s*', '', x))
df_consolidado['Jogadora'] = df_consolidado['Jogadora'].apply(lambda x: re.sub(r'#', ' #', x))

# Extraindo o nome da equipe da coluna "Atleta"
df_consolidado['Equipe'] = df_consolidado['Atleta'].str.extract(r'\((\w+)\)')

# Excluindo a coluna "Atleta"
df_consolidado.drop(columns=['Atleta'], inplace=True)
# Convertendo as colunas relevantes para o tipo numérico
df_consolidado['Pontos'] = pd.to_numeric(df_consolidado['Pontos'], errors='coerce')
df_consolidado['FGA'] = pd.to_numeric(df_consolidado['FGA'], errors='coerce')
df_consolidado['LLT'] = pd.to_numeric(df_consolidado['LLT'], errors='coerce')
df_consolidado['Assistências'] = pd.to_numeric(df_consolidado['Assistências'], errors='coerce')
df_consolidado['Erros'] = pd.to_numeric(df_consolidado['Erros'], errors='coerce')




# Calculando a coluna 'AST/TOV'
df_consolidado['AST/TOV'] = ((df_consolidado['Assistências'] / df_consolidado['Erros'])).round(1)

# Calculando a coluna 'TS%'
df_consolidado['TS%'] = ((df_consolidado['Pontos'] / ((2 * df_consolidado['FGA']) + (0.88 * df_consolidado['LLT']))) * 100).round(1)
df_consolidado['TS%'] = df_consolidado['TS%'].astype(str) + '%'
df_consolidado['PPP'] = (df_consolidado['Pontos'] / (df_consolidado['FGA'] + (0.44 * df_consolidado['LLT']) + df_consolidado['Erros'])).round(2)

# Reorganizando as colunas para incluir 'TS%' ao lado de 'FG%'
colunas_com_ts = ['Jogadora', 'Equipe', 'Jogos', 'Minutoss', 'Pontos', 'PPP', '2FGM', '2FGA', '2FG%', '3FGM', '3FGA', '3FG%', 'FGM', 'FGA', 'FG%', 'TS%', 'LLT', 'LLC', 'LL%', 'FTr',
           'Rebotes', 'Def Reb', 'Off Reb', 'Assistências', 'Erros', 'AST/TOV', 'Roubos', 'Tocos', 'STL + BLK']

df_consolidado = df_consolidado[colunas_com_ts]

# Exibindo o DataFrame consolidado com a nova coluna 'TS%'
print(df_consolidado)
```

```python
# Criando um novo DataFrame dividindo as colunas pelo número de jogos e arredondando para um dígito após a vírgula
df_por_jogo = df_consolidado.copy()
colunas_divididas = ['Minutoss', 'Pontos', '2FGM', '2FGA', '3FGM', '3FGA', 'FGM', 'FGA', 'LLT', 'LLC',
                     'Rebotes', 'Def Reb', 'Off Reb', 'Assistências', 'Erros', 'Roubos', 'Tocos', 'STL + BLK']

# Convertendo as colunas para numérico
df_por_jogo[colunas_divididas] = df_por_jogo[colunas_divididas].apply(pd.to_numeric, errors='coerce')

# Dividindo as colunas pelo número de jogos e arredondando para um dígito após a vírgula
for coluna in colunas_divididas:
    df_por_jogo[coluna] = (df_por_jogo[coluna] / df_por_jogo['Jogos']).round(1)

# Exibindo o novo DataFrame com as colunas divididas pelo número de jogos
print(df_por_jogo)
```

```python
import pandas as pd
import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display, clear_output

# Função para criar o gráfico de radar
def criar_grafico(change):
    # Selecionando apenas as colunas desejadas para o gráfico de radar
    colunas_grafico = ['2FG%', '3FG%', 'FG%', 'TS%', 'LL%']

    # Verificando se as jogadoras estão no DataFrame
    df_comparacao = df_por_jogo[df_por_jogo['Jogadora'].isin([dropdown_jogadora1.value, dropdown_jogadora2.value])]

    if len(df_comparacao) == 0:
        with output:
            print("Uma ou ambas as jogadoras não estão presentes no DataFrame.")
    else:
        # Convertendo as colunas relevantes para números
        for coluna in colunas_grafico:
            df_comparacao[coluna] = pd.to_numeric(df_comparacao[coluna].str.rstrip('%'), errors='coerce')

        # Removendo linhas que possuem valores NaN após a conversão
        df_comparacao.dropna(subset=colunas_grafico, inplace=True)

        if len(df_comparacao) == 0:
            with output:
                print("Os valores das estatísticas são inválidos para criar o gráfico.")
        else:
            # Configurando o layout do gráfico
            fig = go.Figure()

            # Adicionando as linhas para cada jogadora
            for jogador, cor_linha, cor_marca in [(dropdown_jogadora1.value, 'blue', 'darkblue'), (dropdown_jogadora2.value, 'red', 'tomato')]:
                dados_jogadora = df_comparacao[df_comparacao['Jogadora'] == jogador]
                if not dados_jogadora.empty:
                    fig.add_trace(go.Scatterpolar(
                        r=dados_jogadora[colunas_grafico].values[0],
                        theta=colunas_grafico,
                        fill='toself',
                        name=jogador,
                        line=dict(color=cor_linha),
                        marker=dict(color=cor_marca)
                    ))

            # Configurando o layout do gráfico
            fig.update_layout(
                polar=dict(
                    radialaxis=dict(
                        visible=True,
                        tickfont=dict(color='black')
                    ),
                    angularaxis=dict(
                        tickfont=dict(color='black')
                    )
                ),
                legend=dict(
                    orientation="h",
                    yanchor="bottom",
                    y=-0.15,
                    xanchor="center",
                    x=0.5
                ),
                margin=dict(l=50, r=50, t=100, b=50),
                font=dict(family="Arial", size=12),
                width=800,  # Largura do gráfico
                height=600  # Altura do gráfico
            )

            # Exibindo o gráfico
            with output:
                output.clear_output(wait=True)
                display(fig)


# Obtendo a lista de jogadoras disponíveis no DataFrame
jogadoras_disponiveis = df_por_jogo['Jogadora'].unique().tolist()

# Criando os widgets de seleção de jogadora
dropdown_jogadora1 = widgets.Dropdown(options=jogadoras_disponiveis, description='Jogadora 1:')
dropdown_jogadora2 = widgets.Dropdown(options=jogadoras_disponiveis, description='Jogadora 2:')

# Criando o output para exibir o gráfico
output = widgets.Output()

# Observando mudanças nos menus suspensos e chamando a função para atualizar o gráfico
dropdown_jogadora1.observe(criar_grafico, names='value')
dropdown_jogadora2.observe(criar_grafico, names='value')

# Exibindo os widgets
widgets.VBox([dropdown_jogadora1, dropdown_jogadora2, output])
```
